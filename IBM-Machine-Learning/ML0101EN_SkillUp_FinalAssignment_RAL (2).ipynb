### Data Preprocessing


#### One Hot Encoding


First, we need to perform one hot encoding to convert categorical variables to binary variables.


```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

Next, we replace the values of the 'RainTomorrow' column changing them from a categorical column to a binary column. We do not use the `get_dummies` method because we would end up with two columns for 'RainTomorrow' and we do not want, since 'RainTomorrow' is our target.


```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

### Training Data and Test Data


Now, we set our 'features' or x values and our Y or target variable.


```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

```python
features.head()
```

```python
Y.head()
```

```python
#check the shape of the input data X and Y
print(features.shape)
print(Y.shape)
```

### Linear Regression


#### Q1) Use the `train_test_split` function to split the `features` and `Y` dataframes with a `test_size` of `0.2` and the `random_state` set to `10`.


```python
#Enter Your Code and Execute
x_train, x_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=10)
```

#### Q2) Create and train a Linear Regression model called LinearReg using the training data (`x_train`, `y_train`).


```python
#Enter Your Code and Execute
LinearReg = LinearRegression()
LinearReg = LinearReg.fit(x_train, y_train)
```

#### Q3) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


```python
#Enter Your Code and Execute
predictions = LinearReg.predict(x_test)
```

#### Q4) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


```python
#Enter Your Code and Execute
LinearRegression_MAE = metrics.mean_absolute_error(y_test, predictions)
LinearRegression_MSE = metrics.mean_squared_error(y_test, predictions)
LinearRegression_R2 = metrics.r2_score(y_test, predictions)
#LinearR_Accuracy_Score = metrics.accuracy_score(y_test, predictions)
#LinearR_JaccardIndex =  jaccard_score(y_test,predictions)
#LinearR_F1_Score = f1_score(y_test, predictions)
```

#### Q5) Show the MAE, MSE, and R2 in a tabular format using data frame for the linear model.


```python
#Enter Your Code and Execute
# Create a dictionary with the metrics
data = {
    'Metric': ['MAE', 'MSE', 'R2'],
    'Value': [LinearRegression_MAE, LinearRegression_MSE, LinearRegression_R2]
}

# Create a pandas DataFrame from the dictionary
Report = pd.DataFrame(data)
print(Report)
```

### KNN


#### Q6) Create and train a KNN model called KNN using the training data (`x_train`, `y_train`) with the `n_neighbors` parameter set to `4`.


```python
#Enter Your Code and Execute
k = 4
#Train Model and Predict
KNN = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train)
```

#### Q7) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.


```python
#Enter Your Code and Execute
predictions = KNN.predict(x_test)
```

#### Q8) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.


```python
#Enter Your Code and Execute
KNN_Accuracy_Score = metrics.accuracy_score(y_test, predictions)
KNN_JaccardIndex =  jaccard_score(y_test,predictions)
KNN_F1_Score = f1_score(y_test, predictions)
```

```python
# Create a dictionary with the metrics
data = {
    'Metric': ['Accuracy Score', 'Jaccard Index', 'F1 Score'],
    'Value': [KNN_Accuracy_Score, KNN_JaccardIndex, KNN_F1_Score]
}

# Create a pandas DataFrame from the dictionary
Report = pd.DataFrame(data)
print(Report)
```

# Analysis of Results

Let's break down the performance of each model based on the provided metrics: **Accuracy**, **Jaccard Index**, **F1 Score**, and (for Logistic Regression) **Log Loss**.

---

## **1. K-Nearest Neighbors (KNN)**
- **Accuracy**: 0.818 — KNN performs well in terms of classification accuracy, indicating it correctly classifies ~82% of the instances.
- **Jaccard Index**: 0.425 — The Jaccard Index, measuring the similarity between predicted and actual labels, is moderate, suggesting room for improvement in overlapping predictions.
- **F1 Score**: 0.597 — This moderate value indicates that precision and recall are reasonably balanced but not optimal.
- **Log Loss**: Not Applicable (NA) — KNN does not provide probabilistic outputs, so Log Loss isn't computed.

**Comment**: KNN shows good accuracy but struggles with overlap and balancing precision and recall. It might benefit from parameter tuning (e.g., adjusting the number of neighbors).

---

## **2. Decision Trees**
- **Accuracy**: 0.818 — Same as KNN, showing comparable accuracy in classification.
- **Jaccard Index**: 0.480 — Higher than KNN, meaning Decision Trees produce predictions that better match the actual labels.
- **F1 Score**: 0.649 — The best among all models so far, indicating a good balance of precision and recall.
- **Log Loss**: NA — Decision Trees don't output probabilities in a manner suitable for Log Loss.

**Comment**: Decision Trees outperform KNN in the Jaccard Index and F1 Score. They seem to capture the structure of the data better. However, they may be prone to overfitting, so regularization (e.g., pruning) should be considered.

---

## **3. Logistic Regression**
- **Accuracy**: 0.827 — The best accuracy among all models, suggesting it generalizes well to the data.
- **Jaccard Index**: 0.484 — Slightly better than Decision Trees, showing good alignment with actual labels.
- **F1 Score**: 0.652 — Marginally better than Decision Trees, indicating good precision and recall balance.
- **Log Loss**: 0.380 — This low value indicates that Logistic Regression provides well-calibrated probabilities.

**Comment**: Logistic Regression is the best-performing model overall, with the highest accuracy, Jaccard Index, and F1 Score, as well as a low Log Loss. It is likely a good candidate for deployment.

---

## **4. Support Vector Machines (SVM)**
- **Accuracy**: 0.722 — The lowest accuracy among all models, suggesting poor generalization to the data.
- **Jaccard Index**: 0.000 — Indicates no overlap between predicted and actual labels.
- **F1 Score**: 0.000 — Indicates the model failed to classify any instances correctly in terms of true positives.
- **Log Loss**: NA — Log Loss isn't provided for SVM.

**Comment**: SVM performs poorly, likely due to inappropriate parameter selection (e.g., kernel type or regularization strength) or an inability to separate the classes effectively given the feature space.

---

## **Overall Recommendations**
1. **Best Model**: Logistic Regression shows the most consistent and reliable performance across all metrics. It should be prioritized for further tuning or deployment.
2. **Secondary Option**: Decision Trees perform almost as well as Logistic Regression and might be preferable if interpretability is crucial.
3. **KNN**: While accurate, it struggles with precision and recall balance. Parameter tuning or feature scaling might improve its performance.
